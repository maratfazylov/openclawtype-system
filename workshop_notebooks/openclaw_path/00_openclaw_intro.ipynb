# 00. OpenClaw: вводная презентация без кода

Этот notebook задаёт историю для workshop. Здесь нет code cells: только тезисы, схемы и порядок разговора перед интерактивной сборкой.


## 1. OpenClaw и его величие

![OpenClaw story](../visuals/openclaw_path/00_openclaw_story.svg)

OpenClaw удобно объяснять как правильный UX для агента: пользователь пишет в привычный канал, а за сообщением стоит runtime, который умеет думать, вызывать tools, держать state и возвращать результат.

Главная ценность не в «ещё одном чате», а в том, что мессенджер становится входом в рабочий контур.


## 2. Косяки и безопасность

![Security failures](../visuals/openclaw_path/01_security_failures.svg)

У такого UX сразу появляется цена: агент действует на основе внешнего текста. Значит, `messages` — это не просто интерфейс, а граница доверия.

Что важно проговорить:

- prompt injection приходит как обычный пользовательский или web-контент;
- tool misuse случается, когда нет dry-run, approval gate и scoped credentials;
- утечки появляются, если агент смешивает секреты, logs и внешний ответ.


## 3. Начинаем с API: messages

Минимальная форма агента — это список сообщений:

```json
{
  "messages": [
    {"role": "user", "content": "Проверь job и скажи статус"}
  ]
}
```

От этой формы удобно танцевать дальше: LangChain вызывает модель, LangGraph удерживает state, Deep Agents добавляют files, planning, tools, subagents и production-поведение.


## 4. LangChain → LangGraph → Deep Agents

![LangChain ladder](../visuals/openclaw_path/02_langchain_ladder.svg)

Переходы простые:

- LangChain: модель и tool-calling;
- LangGraph: state machine, edges, checkpointing;
- Deep Agents: готовый harness для длинных задач, файлов, планирования и делегирования.


## 5. Что такое Deep Agent

![Deep Agent anatomy](../visuals/openclaw_path/03_deep_agent_anatomy.svg)

Deep Agent — это runtime вокруг модели: инструкции, tools, файловый backend, planning loop, subagents и checkpoints. Через него удобно объяснить, почему Claude Code ощущается не как чат, а как инженерный агент.

Отсюда естественный переход к SWE: планирование, работа с файлами, локализация проблемы, патч, тесты.
